```python
#what category does each animal fit under?
humans =  [x.upper() for x in ['HUMAN ON DONKEY','HUMAN ON HORSE','VEHICLE','VEHICLES/HUMANS/LIVESTOCK','HUMAN ON FOOT', 'CAMERA TRAP TEAM']]
domestic_list = [x.upper() for x in['COW','DONKEY','GOAT','SHEEP','HORSE']]
wild_herb = [x.upper() for x in['DUIKER', 'ELEPHANT', 'GEMSBOK', 'GIRAFFE', 'KUDU', 'OSTRICH', 'PORCUPINE', 'SCRUB HARE', 'SPRINGBOK', 'SPRINGHARE', 'STEENBOK', 'WILDEBEEST', 'ZEBRA']]

def label (row):
    if row['Species'] in domestic_list:
        return 'domestic'
    if row['Species'] in humans:
        return 'human'
    if row['Species'] in wild_herb:
        return 'wild'
    return 'NA'

df_species['Type'] = df_species.apply(lambda row: label(row), axis=1)
```

```python
df_species.head()
```

```python
#df_species.drop(df_species.Type == 'NA', inplace = 'True')

df_studyspecies= df_species.loc[(df_species.Type == 'domestic')| (df_species.Type == 'human')| (df_species.Type == 'wild')]
df_studyspecies
```

##Including Distance to Variables

```python
#import station folder w/ relative distance to each variable

file = '/content/stations_Feb2023.csv'
df_stations= pd.read_csv(file)
df_stations.head()
```

```python
df_stations.info()
```

Here, we can see we have a Near_road column as well as a road_m column. Either these columns repeat and one needs to be dropped, or they mean different things. Based on df_stations.head(), we see that Near_road and road_m don't match up, so they must mean different things. The "near___" columns mean distance the location is to that object, but I don't have much of an idea of what road_m means. 

```python
#convert distances into km

df_stations['Dist_CP_km']= df_stations['Near_CP']/1000
df_stations['Dist_road_km']= df_stations['Near_road']/1000
df_stations['Dist_river_km']= df_stations['Near_river']/1000
df_stations['Dist_village_km']= df_stations['Near_Vil']/1000
```

```python
df_stations.head()
```

```python
#melt table to turn distance to river, road, CP, and village into variables
df_distances= pd.melt(df_stations, id_vars=['Deployment.Location.ID'], value_vars=['Dist_river_km', 'Dist_road_km', 'Dist_CP_km', 'Dist_village_km'])
df_distances
```

```python
plotnine.options.figure_size = (25,10)
(
    ggplot(df_distances, mapping = aes(x='Deployment.Location.ID', y='value', fill='variable'))
    + geom_point(size = 5)
    + theme(axis_line=element_line(color="gray"), 
            axis_ticks=element_line(color = "black"), 
            text=element_text(size = 12),
            axis_text_x=element_text(rotation=90))
)
```

```python
plotnine.options.figure_size = (25,10)
(
    ggplot(df_distances)
    + geom_bar(stat= 'identity', mapping = aes(x='Deployment.Location.ID', y='value', fill='variable'), width = 1, position= 'dodge')
    + theme(axis_line=element_line(color="gray"), 
            axis_ticks=element_line(color = "black"), 
            text=element_text(size = 12),
            axis_text_x=element_text(rotation=90))
    + xlab('Deployment Location ID')
    + ylab('Distance (km)')
    + ggtitle('Distance of Each Camera to Certain Landscape Markers')
)
```

**Is there a relationship between number of wild herbivores and distance to river?**

Need to merge station data with species dataframe. Then need to sort out the different types and visualize data.

```python
df_n_type= df_studyspecies.groupby(['Deployment.Location.ID', 'Type']).agg(total = ('Number.of.Animals', 'sum')).reset_index()
df_n_type
```

```python
df_n_type_season= df_studyspecies.groupby(['Deployment.Location.ID', 'Type', 'Season']).agg(total = ('Number.of.Animals', 'sum')).reset_index()
df_n_type_season.head()
```

```python
df_infos = pd.merge(df_n_type_season, df_stations, how="left", on= ["Deployment.Location.ID", 'Season'])
df_infos.head()
```

```python
plotnine.options.figure_size=(8,5)

(
    ggplot(df_infos, aes(x='Dist_riv_km', y='total'))
    + geom_point(aes(x='total', y='Dist_river_km', color='Season'))
    + facet_wrap(facets='Type')
    + xlab('Distance to River (km)')
    + ylab('Number of Animals')
    + ggtitle('Animal Locations Relative to River')
)
```

This shows us that wild animals are less likely to steer away from water than domestic animals. 

Let's see if this trend is similar for the 6 study sites. 

Now, we're going to use our six selected deployment locations (Y6500, Y6200, Y0004, Y0034, Y7700, and Y8100) and compare the number of animals found to four different variables (dist to road, river, village, cattle post). 

```python
df_studysite= df_n_type_season[(df_n_type_season['Deployment.Location.ID']== 'Y6500')| (df_n_type_season['Deployment.Location.ID']== 'Y6200')|(df_n_type_season['Deployment.Location.ID']== 'Y0004')|(df_n_type_season['Deployment.Location.ID']== 'Y0034')|(df_n_type_season['Deployment.Location.ID']== 'Y7700')|(df_n_type_season['Deployment.Location.ID']== 'Y7700')|(df_n_type_season['Deployment.Location.ID']== 'Y8100')]
df_studysite.head()
```

```python
df_studysite_infos = pd.merge(df_studysite, df_stations, how="left", on= ["Deployment.Location.ID", 'Season'])
df_studysite_infos.head()
```

```python
plotnine.options.figure_size=(8,5)

(
    ggplot(df_studysite_infos, aes(x='Dist_riv_km', y='total'))
    + geom_point(aes(x='total', y='Dist_river_km', color='Season'))
    + facet_wrap(facets='Type')
    + xlab('Distance to River (km)')
    + ylab('Number of Animals')
    + ggtitle('Animal Locations Relative to River')
)
```

This chart also shows us that the season doesn't really impact an animals relative location to water sources.
